## 1.1 THEORY
### Data Wrangling
##### Data wrangling is the process of converting raw data into a usable form. It may also be called data munging or data remediation. It describes a series of processes designed to explore, transform, and validate raw datasets from their messy and complex forms into high-quality data. We can use our wrangled data to produce valuable insights and guide business decisions. 
##### Data wrangling steps are systematic stages in transforming raw data into a structured, clean and usable format for analysis. These steps ensure accurate and actionable insights for decision making by addressing raw data's challenges and improving it's usability.
#### Data Wrangling Steps
##### **1. Discovery** 
##### Discovery refers to the process of familiarizing with data so we can conceptualize how we might use it. We can liken it to looking in our refrigerator before cooking a meal to see what ingredients we have at our disposal.During discovery, we may identify trends or patterns in the data, along with obvious issues, such as missing or incomplete values that need to be addressed. This is an important step, as it will inform every activity that comes afterward.
##### **2. Structuring**
##### Raw data is typically unusable in its raw state because itâ€™s either incomplete or misformatted for its intended application. Data structuring is the process of taking raw data and transforming it to be more readily leveraged. The form our data takes will depend on the analytical model we use to interpret it.
##### **3. Cleaning**
##### Data cleaning is the process of removing inherent errors in data that might distort the analysis or render it less valuable. Cleaning can come in different forms, including deleting empty cells or rows, removing outliers, and standardizing inputs. The goal of data cleaning is to ensure there are no errors (or as few as possible) that could influence the final analysis. Identifying and removing any bad data greatly impacts the rest of the wrangling processes.
##### **4. Enriching**
##### Enrichment involves adding value to our data by incorporating additional context or merging it with other relevant datasets. This can include combining data from various sources to create new variables or applying domain-specific transformations. For eg: Enrichment could involve segmenting customers into a dataset based on their purchasing behaviour.
##### **5. Validation**
##### Data validation involves ensuring the accuracy and reliability of your dataset through systematic checks and automated processes. This step is critical for confirming that data is both correct and consistent. Its goal is to catch any issues before the data is used for analysis, preventing errors from impacting the results.
##### **6. Data Publishing**
##### The final step of data wrangling involves storing the cleared and structured data in a readily accessible format for analysis. This may involve storing data in a database, data warehouse or analytics platform. It also involves documenting data with metadata detailing its content, structure and any transformation it has undergone.

### Data Transformation and Normalization Techniques
##### **Min-Max Transformation**
##### The main purpose of this transformation is it is used when the data has no outliers and if we want to scale it to a specific range[0, 1] for algorithms sensitive to scale, such as K-Nearest Neighbours(K-NN) and Neural Networks.
##### **Z-Score Normalization**
##### It is used when the data has outliers or when features have different units. Suitable for algorithms like sum and logistic regression.

## Exploratory Data Analysis(EDA)
##### Exploratory Data Analysis (EDA) is the initial step in the data analysis process where analysts use statistical and visualization techniques to understand the main characteristics of a dataset. The goal is to summarize its key features, discover patterns, identify anomalies, check assumptions, and test hypotheses. EDA is often used before applying machine learning models or other advanced analytical techniques to ensure the data is clean, well-understood, and ready for further analysis.
 ### Key Steps in Exploratory Data Analysis
 ##### **1. Understand the Data Structure:**
 - Data types: Identify and inspect data types for each feature (numerical, categorical, datetime, etc.).
 - Basic summary statistics: Calculate key metrics like mean, median, mode, standard deviation, and percentiles
 - Data shape: Check the number of rows (observations) and columns (features) in the dataset.
##### **2. Data CLeaning:**
- Data cleaning is the process of removing inherent errors in data that might distort the analysis or render it less valuable.
-  Cleaning can come in different forms, including deleting empty cells or rows, removing outliers, and standardizing inputs.
##### **3. Univariate Analysis (Single Variable):**
- Histograms/Bar plots: Visualize the distribution of a single feature. For continuous variables, histograms are useful, while for categorical data, bar plots can be used.
- Box plots: Used to detect outliers and understand the spread of data. For numerical features, they show the median, quartiles, and outliers.
- Summary statistics: Review the mean, median, mode, range, and standard deviation to understand central tendencies and variability.
##### **4. Bivariate Analysis (Two Variables):**
- Scatter plots: Used to explore relationships between two continuous variables, especially for detecting trends or correlations.
- Correlation matrix: For numerical features, calculating correlation coefficients (like Pearson or Spearman) can help identify relationships between variables.
 ##### **5. Multivariate Analysis (Multiple Variables):**
 - Heatmaps: A heatmap of a correlation matrix can provide a compact view of how multiple features are correlated.
 - Principal Component Analysis (PCA): PCA is used for dimensionality reduction to find underlying patterns in data and visualize high-dimensional data in fewer dimensions.
 - Cluster analysis: Grouping similar data points together using clustering algorithms like k-means or hierarchical clustering.
##### **6. Identify Outliers and Anomalies:**
- Use statistical methods (like Z-scores) or visualization techniques (box plots, scatter plots) to detect outliers or anomalous data points that may need to be addressed.
##### **7. Feature Engineering:**
- Generate new features that might help in analysis or model performance (e.g., combining existing features, transforming data types, etc.).
##### **8. Data Visualization:**
- Histograms, bar charts, box plots: For univariate analysis.
- Scatter plots, line plots: For bivariate relationships.
- Heatmaps, pair plots: For multivariate exploration.
- Visualizing your data helps identify patterns and potential relationships that might not be obvious from just summary statistics.
##### **9.Testing Assumptions:**
- Based on the data, test assumptions that might be relevant for future analyses (e.g., checking if data is normally distributed, verifying if certain relationships hold true across different variables).

### Importance of Exploratory Data Analysis(EDA)
##### **Understanding the Data:**
- Before diving into more advanced analysis, EDA allows you to understand the underlying structure of your data, the relationships between variables, and the distribution of your data.

##### **Data Quality Checks:**
- EDA helps identify data quality issues such as missing values, outliers, duplicates, or incorrect data types. Addressing these issues is crucial for accurate analysis.

##### **Identifying Patterns and Trends:**
- Through visualization and statistical analysis, EDA reveals patterns, correlations, and potential relationships that can inform modeling or decision-making.

##### **Hypothesis Generation:**
- By exploring the data, you can generate hypotheses that can later be tested with more formal statistical techniques or machine learning models.

##### **Feature Selection:**
- EDA helps in identifying the most important features for modeling by evaluating their relevance and importance. You may discard irrelevant or redundant features at this stage.

### Feature Scaling(Data Normalization)
##### Z-score Standardization used to scale the features such that they have a mean of 0 and a standard deviation of 1. Min-Max Scaling transforms the numeric features into a fixed range, typically between 0 and 1.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Z-score standardization
scaler = StandardScaler()
# Exclude price column for standardization
new_numeric_columns = [col for col in numeric_columns if col != "price"]
df_zscore = pd.DataFrame(scaler.fit_transform(df[new_numeric_columns]), columns=new_numeric_columns)
df[new_numeric_columns] = df_zscore

# Min-Max scaling
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df[numeric_columns]), columns=numeric_columns)
df[numeric_columns] = df_minmax
```

### Categorical Variable Encoding
##### It is applied to categorical columns, transforming each category (or unique value) into a corresponding numeric label.

```python
# Replace NAN in numeric columns with the mean, and in the categorical columns with the mode
df[numeric_columns] = df[numeric_columns].apply(lambda col: col.fillna(col.mean()))
```

# HANDLING OUTLIERS

```python
df.shape
```

```python
# remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

for column in numeric_columns:
    df = remove_outliers(df, column)
    
```
